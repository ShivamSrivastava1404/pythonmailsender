```python
from tkinter import *
from email.mime.multipart import MIMEMultipart #object of a standard email
from email.mime.base import MIMEBase #object for attachments of a email
from email.mime.text import MIMEText # object to add text to our email
import email
from tkinter import messagebox
from tkinter import filedialog






scr=Tk(className='Gmail')
scr.geometry('600x800')
scr.config(bg='lightgray')
a=Label(scr,text='Welcome! to....',bg='black',fg='red',font=('times',25,'bold','underline'),relief='raised')
a.pack(side=TOP,fill=X)

img=PhotoImage(file='img2.png')
p=Label(scr,image=img)
p.image=img
p.pack()

mail=Label(scr,text='Your Email Account',font=('times',15,'bold'),width=20)
mail.place(x=20,y=160)
_mail=Entry(scr,width=45)
_mail.place(x=300,y=163)



paswd=Label(scr,text='Your Email Password',font=('times',15,'bold'),width=20)
paswd.place(x=20,y=210)
_paswd=Entry(scr,width=45,show='*')
_paswd.place(x=300,y=213)


to=Label(scr,text='To',font=('times',15,'bold'),width=20)
to.place(x=20,y=260)
_to=Entry(scr,width=45)
_to.place(x=300,y=263)

sub=Label(scr,text='Subject',font=('times',15,'bold'),width=20)
sub.place(x=20,y=310)
_sub=Entry(scr,width=45)
_sub.place(x=300,y=313)



msg=Label(scr,text='Body',font=('times',15,'bold'),width=20)
msg.place(x=20,y=360)
_msg=Text(scr,width=34,height=10)
_msg.place(x=300,y=363)

def fun():
    _attach.insert(0,filedialog.askopenfilename())
    f=_attach.get()
    s=""
    for i in f[::-1]:
        if i=="/":
            break
        else:
            s+=i
    r=s[::-1]
    _img.insert(0,r)
    

attach=Label(scr,text="Attachment's",font=('times',15,'bold'),width=20)
attach.place(x=20,y=560)
choose=Button(scr,text='Choose file',bg="black",fg="red",width=12,command=fun)
choose.place(x=300,y=563)
_attach=Entry(scr,width=29)
_attach.place(x=395,y=563)
_img=Entry(scr)



def send():
    
    em=MIMEMultipart()
    em['from']=_mail.get()
    em['to']=_to.get()
    em['subject']=_sub.get()
    
    em.attach(MIMEText(_msg.get(1.0,'end'),'plain'))
    
    at=MIMEBase('application','octet-stream')#attachment can be a apllication of octet data steram
    at.set_payload(open(_attach.get(),'rb').read())#to upload the file
    email.encoders.encode_base64(at) # to encode the data to the 64bit format
    at.add_header('content-disposition','attachment;filename={}'.format(_img.get()))
    em.attach(at)
    
    import smtplib
    server=smtplib.SMTP('smtp.gmail.com',587)
    server.starttls()
    server.login(_mail.get(),_paswd.get())
    server.sendmail(_mail.get(),_to.get(),em.as_string())
    
    messagebox.showinfo('Gmail','Email sent successfully!')
    
    
def clear():
    _mail.delete(0,END)
    _paswd.delete(0,END)
    _to.delete(0,END)
    _sub.delete(0,END)
    _msg.delete('1.0',END)
    _attach.delete(0,END)
    _img.delete(0,END)
    messagebox.showinfo('Gmail','All entry clear successfully!')
    

    
cancil=Button(scr,text='clear',activebackground="red",activeforeground="black",
         bd=10,bg="black",fg="red",relief='sunken',width=18,command=clear)
cancil.place(x=100,y=663)

i=Button(scr,text='send',activebackground="green",activeforeground="white",
         bd=10,bg="black",fg="red",relief='sunken',width=18,command=send)
i.place(x=400,y=663)

scr.mainloop()
```
